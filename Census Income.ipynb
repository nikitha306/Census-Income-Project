```python
for col in data.columns:
    if data[col].dtype!='object':
        plt.boxplot(data[col])
        plt.xlabel(col)
        plt.show()
```

```python
data['capital-gain'].value_counts()
```

```python
out_col=['hours-per-week','fnlwgt','age','education-num']
```

```python
# Outlier removal
for col in out_col:
    Q1=data[col].quantile(0.25)
    Q3=data[col].quantile(0.75)
    IQR=Q3-Q1
    LW=Q1-1.5*IQR  #lower Wiskers
    UW=Q3+1.5*IQR #upper wiskers
    data=data[(data[col]>=LW) &(data[col]<=UW)]    
```

```python
data
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for col in data.columns:
    if data[col].dtype=="object":
        data[col]=le.fit_transform(data[col])  # labels will be assigned in alphabetical order
```

```python
#Split data into training and testing
from sklearn.model_selection import train_test_split
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lo_model=LogisticRegression()
```
